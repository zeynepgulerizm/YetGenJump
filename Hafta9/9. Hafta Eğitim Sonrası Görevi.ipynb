# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda():
    def __init__(self,tv_durum="Kapalı", tv_ses=0, kanal_listesi=["TRT"],kanal="TRT"):
        self.tv_durum= tv_durum
        self.tv_ses= tv_ses
        self.kanal_listesi= kanal_listesi
        self.kanal= kanal

    def tv_aç(self):
        if (self.tv_durum== "Kapalı"):
            print("TV açılıyor...")
            self.tv_durum="Açık"
            time.sleep(1)
            print("TV açıldı.")
        else:
            print("TV zaten açık.")

    def tv_kapat(self):
        if (self.tv_durum == "Açık"):
            print("TV kapatılıyor...")
            self.tv_durum = "Kapalı"
            time.sleep(1)
            print("TV kapandı.")
        else:
            print("TV zaten kapalı.")

    def ses_ayarla(self):
        while True:
            işlem= input("Ses arttır(+)\nSes azalt(-)\nÇıkış yap(q)\İşlem:  ")
            if işlem=="+" and self.tv_ses<=30:

                self.tv_ses+=1
                print("Ses düzeyi ", self.tv_ses)

            elif işlem=="-" and self.tv_ses>0:

                self.tv_ses-=1
                print("Ses düzeyi ", self.tv_ses)

            elif işlem=="q" or işlem=="Q":
                print("Ses düzeyi ", self.tv_ses)
                time.sleep(1)
                break
            else:
                print("Ses düzeyi ", self.tv_ses)


    def kanal_ekle(self, yeni_kanal):

        if yeni_kanal not in self.kanal_listesi:
            self.kanal_listesi.append(yeni_kanal)

            print("Kanal eklendi.")
            time.sleep(1)
        else:

            print("Kanal zaten mevcut")
            time.sleep(1)

    def rastgele_kanal(self):
        rastgele = random.randint(0, len(self.kanal_listesi)-1)
        self.kanal= self.kanal_listesi[rastgele]
        print("Açık olan kanal ",self.kanal)

    def __len__(self):
        return len(self.kanal_listesi)

    def __str__(self):

        return ("TV Durumu: {}\nSes Düzeyi: {}\nKanal Listesi: {}\nAçık olan kanal: {}".format(self.tv_durum,self.tv_ses,self.kanal_listesi,self.kanal))


    def kanal_silme(self, silinecek):
        self.kanal_listesi.remove(silinecek)
        print("Kanal silindi.")

kumanda=Kumanda()


print("""
    Kumanda Uygulaması
1. TV Açma
2. TV Kapama
3. Ses ayarları
4. Kanal Ekle
5. Kanal Sil
6. Rastgele Kanala Geç
7. Current Information
8. Açık olan kanal
Çıkmak için 'q' ya basın.
"""
)

while True:
    işlem2= input("Yapmak istediğiniz işlemin numarasını giriniz: ")
    if işlem2=="1":
        kumanda.tv_aç()
    elif işlem2=="2":
        kumanda.tv_kapat()
    elif işlem2=="3":
        kumanda.ses_ayarla()
    elif işlem2=="4":
        eklenecekler= input("Eklenecek kanallar arasına virgül koyunuz: ")
        kanal_listesi= eklenecekler.split(",")

        for i in kanal_listesi:
            kumanda.kanal_ekle(i)

    elif işlem2=="5":
        sil=input("Silinecek kanalın adını girin: ")
        kumanda.kanal_silme(sil)
    elif işlem2=="6":
        kumanda.rastgele_kanal()
    elif işlem2=="7":
        print(kumanda)
    elif işlem2=="8":
        print("Açık olan kanal " ,kumanda.kanal)
    elif işlem2=="q" or işlem2== "Q":
        print("Kumanda Kapatılıyor.")
        time.sleep(2)
        break
    else:
        print("Geçersiz işlem.")
```
